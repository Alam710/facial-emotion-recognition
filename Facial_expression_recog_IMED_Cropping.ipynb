```python
def fileCount(folder, allowed_extensions=None):
   "count the number of files in a directory and sub directory"
   count = 0
   for base, dirs, files in os.walk(folder):
      for file in files:
         if allowed_extensions and file.endswith(allowed_extensions) or not allowed_extensions:
            count += 1
   return count
```

## Crop Image

```python
from sys import maxsize
import cv2
from google.colab.patches import cv2_imshow
from os import path
from tqdm import tqdm

# select model
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
# create dir
if path.exists('IMED') == False:
  os.makedirs('IMED')
# get all image
id = 1
for i in tqdm(range(len(file_paths))):
  get_folder_name = file_paths[i].split('/')
  get_img = cv2.imread(file_paths[i])
  gray = cv2.cvtColor(get_img, cv2.COLOR_RGB2GRAY)

  # detect and crop
  faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.3,
    minNeighbors=6,
    minSize=(30, 30),
  )

  # print("[INFO] Found {0} Faces!".format(len(faces)))

  for (x, y, w, h) in faces:
    cv2.rectangle(gray, (x, y), (x + w, y + h), (0, 255, 0), 2)
    roi_color = gray[y:y + h, x:x + w]

  # save img
  save_path = '/content/IMED/' + get_folder_name[3]
  if path.exists(save_path) == False:
    os.mkdir(save_path)
  cv2.imwrite(save_path + '/' + str(w) + str(h) + str(i) + '.jpg', roi_color)
  # print(f"Image saved to {save_path}")
    # cv2.imwrite(save_path, roi_color)
  id += 1
```

```python
def print_time_taken(start_time, end_time):
    secs_elapsed = end_time - start_time

    SECS_PER_MIN = 60
    SECS_PER_HR  = 60 * SECS_PER_MIN

    hrs_elapsed, secs_elapsed = divmod(secs_elapsed, SECS_PER_HR)
    mins_elapsed, secs_elapsed = divmod(secs_elapsed, SECS_PER_MIN)

    if hrs_elapsed > 0:
        print('Time taken: %d hrs %d mins %d secs' % (hrs_elapsed, mins_elapsed, secs_elapsed))
    elif mins_elapsed > 0:
        print('Time taken: %d mins %d secs' % (mins_elapsed, secs_elapsed))
    elif secs_elapsed > 1:
        print('Time taken: %d secs' % (secs_elapsed))
    else:
        print('Time taken - less than 1 sec')
```
